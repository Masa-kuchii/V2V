```python
%matplotlib inline
import pandas
from pandas import read_csv
import os
import matplotlib.pyplot as plt
import copy
import seaborn as sns
from scipy import stats
import numpy as np

x = [95.2, 95.9, 93.6, 93.2, 94.4, 93.9, 95.4, 93.6, 95.1, 93.7]
y = [94.9, 93.4, 91.3, 89.4]

# x = [95.11, 95.01, 96.19, 95.09, 93.86, 94.88, 94.45, 95.03, 94.72, 93.58, 92.33, 94.44, 93.49]
# y = [93.50, 92.86, 95.69, 89.75, 94.36, 95.29]

nx = len(x)
ny = len(y)

xmean = sum(x)/nx
ymean = sum(y)/ny

x_mean = [xmean]*nx
y_mean = [ymean]*nx

plt.plot(x, linestyle='None', label="Simulation result", color="red", marker="o")
plt.plot(x_mean, linestyle="dashed", label="Simulation mean", color="red")
plt.plot(y, linestyle='None', label="Measurements", color="blue", marker="o")
plt.plot(y_mean, linestyle="dashed", label="Measurement mean", color="blue")


plt.ylabel('Mean speed [km/hour]')
plt.legend()
plt.show()
```

```python
xx = np.array(x)
yy = np.array(y)
res = stats.ttest_ind(xx, yy)
res
```
